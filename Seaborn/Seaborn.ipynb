```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
f = sns.load_dataset("flights") 
```

```python
f.info()
```

```python
f.corr()
```

```python
f.head()
```

```python
#relplot

sns.relplot(x="passengers", y="month", data=f)
```

```python
#relplot2

sns.relplot(x="passengers", y="year", data=f)
```

```python
#so we see over the times passengers are increasing
```

```python
sns.relplot(x="passengers", y="month", hue="year", data=f)
```

```python
sns.relplot(x="passengers", y="year", hue="month", data=f)
```

```python
sns.set(style="ticks")
data = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=data)
```

```python
data.head()
```

```python
sns.catplot(x="day", y="total_bill", kind="violin", data=data)
```
